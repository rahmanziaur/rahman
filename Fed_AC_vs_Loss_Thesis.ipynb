```python

#%% [code]
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import make_interp_spline, BSpline

df1 = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/rahman/master/Train_Accuracy.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/rahman/master/Validation_Accuracy.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/rahman/master/Train_Loss.csv')
df4 = pd.read_csv('https://raw.githubusercontent.com/rahmanziaur/rahman/master/Validation_Loss.csv')


fig, ax1 = plt.subplots()


x1 = df1['Epoch']
y1 = df1['Train_Accuracy']

xnew1 = np.linspace(x1.min(), x1.max(), 300) 
spl1 = make_interp_spline(x1, y1, k=3) 
power_smooth1 = spl1(xnew1)


l1 = ax1.plot(xnew1, power_smooth1, '-', color='#0033CC',label='Train Accuracy')
ax1.legend(loc='upper right', bbox_to_anchor=(0.902, 0.8))

ax1.text(15, 0.75, 'Accuracy', va='center', rotation='vertical')
ax1.text(88, 0.88, 'Epoch', ha='center')


ax1.tick_params(axis='y')
ax1.axis([0, 200, 0.70, 0.90])



ax2 = ax1.twiny() 



x2 = df2['Epoch']
y2 = df2['Validation_Accuracy']

xnew2 = np.linspace(x2.min(), x2.max(), 300) 
sp2 = make_interp_spline(x2, y2, k=3) 
power_smooth2 = sp2(xnew2)

l2 = ax2.plot(xnew1, power_smooth2, '-', color='purple',label='Validation Accuracy')
ax2.legend(loc='upper right', bbox_to_anchor=(0.99, 0.7))

ax2.text(-35, 0.80, '', va='center', rotation='vertical')
#ax2.text(500, 0.88, 'Number of Epoch', ha='center')

ax2.tick_params(axis='y')
ax2.axis([0, 200, 0.70, 0.90])



ax3 = ax2.twinx() 



x3 = df3['Epoch']
y3 = df3['Train_Loss']

xnew3 = np.linspace(x3.min(), x3.max(), 300) 
sp3 = make_interp_spline(x3, y3, k=3) 
power_smooth3 = sp3(xnew3)


l3 = ax3.plot(xnew3, power_smooth3, '-', color='green', label='Train Loss')
ax3.legend(loc='upper right', bbox_to_anchor=(0.825, 0.6))

ax3.text(185, 0.30, 'Loss', va='center', rotation='vertical')
#ax3.text(500, 0.88, 'Number of Epoch', ha='center')

ax3.tick_params(axis='x')
ax3.axis([0, 200, 0.25, 0.50])



ax4 = ax3.twiny() 


x4 = df4['Epoch']
y4 = df4['Validation_Loss']


xnew4 = np.linspace(x4.min(), x4.max(), 300) 
sp4 = make_interp_spline(x4, y4, k=3) 
power_smooth4 = sp4(xnew4)


l4 = ax4.plot(xnew4, power_smooth4, '-', color='brown', label='Validation Loss')
ax4.legend(loc='upper right', bbox_to_anchor=(0.91, 0.5))

ax4.text(225, 0.38, '', va='center', rotation='vertical')
ax4.text(88, 0.26, 'Epoch', ha='center')

ax4.tick_params(axis='x')
ax4.axis([0, 200, 0.25, 0.50])


ax1.grid(b=True, color='#999999', linestyle='--')
ax3.grid(b=True, color='#999999', linestyle=':')


fig.tight_layout()

plt.grid(b=True, which='major', color='#999999', linestyle='--')
fig.savefig('validationAccuracyLoss.pdf')
plt.show()
```
